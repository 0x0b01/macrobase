```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
size = 10_000
d1 = np.random.randint(low=0,high=5,size=size)
d2 = np.random.randint(low=0,high=5,size=size)
xs1 = np.random.normal(loc=50,scale=5,size=size)
ts = np.concatenate([np.repeat(0,size//2), np.repeat(1,size//2)])
```

```python
df = pd.DataFrame({
    "t": ts,
    "d1": d1,
    "d2": d2,
    "x": xs1
})
g1 = (df["d1"] == 0) & (df["d2"] == 0) & (df["t"] == 1)
ng1 = np.sum(g1)
df.loc[g1,"x"] = np.random.normal(loc=300,scale=30,size=ng1)
```

```python
df.groupby(["t","d1"]).aggregate({
    "x": ["count", "mean", "std"]
})
```

```python
df.to_csv("synth_cluster_tight.csv", index=False)
```
